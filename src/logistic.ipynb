```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
f = pd.read_csv('us_oil.csv')
f
```

```python
plt.figure(figsize=(8, 8))
plt.plot(f['volume'])
plt.show()
```

```python
F = []
F.append(f['volume'][0])
for i in range(len(f) - 1):
	F.append(F[i] + f['volume'][i + 1])
```

```python
plt.figure(figsize=(8, 8))
plt.plot(F)
plt.show()
```

```python
x = np.array(F[71:151])
y = f['volume'][71:151] / F[71:151]
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
```

```python
plt.figure(figsize=(8, 8))
plt.plot(x, p(x))
plt.plot(x, y, 'rx')
plt.show()
```
